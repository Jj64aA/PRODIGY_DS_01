## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```python
dataset = pd.read_excel('Population_total_1960_to_2023.xlsx')
```

```python
dataset.head()
print(dataset.columns)
```

## Census extraction in 1960 and determines top 10 and low 10 

```python
population_1960 = dataset[['Country Name', 1960]]
top_10_countries_1960 = population_1960.sort_values(by=1960, ascending=False).head(10)
low_10_countries_1960 = population_1960.sort_values(by=1960 , ascending=True).head(10)
```

```python
top_10_countries_1960
```

## Plotting the top 10 countries and lowest 10 countries (1960)

```python
# top 10 countries
plt.figure(figsize=(10, 6))
plt.barh(top_10_countries_1960['Country Name'], top_10_countries_1960[1960], color='green')
plt.xlabel('Population in 1960')
plt.title('Top 10 Countries by Population in 1960')
plt.gca().invert_yaxis()  
plt.show()

# lowest 10 countries
plt.figure(figsize=(10, 6))
plt.barh(low_10_countries_1960['Country Name'], low_10_countries_1960[1960], color='red')
plt.xlabel('Population in 1960')
plt.title('Lowest 10 Countries by Population in 1960')
plt.gca().invert_yaxis() 
plt.show()
```

# Representation of 5 low and 5 top countries in 5 different stages

### Top 5 Countries by 1960 1980 2000 2020 's population

```python
def top_countries(list_years) :
    for year in list_years : 
        top_10 = dataset[['Country Name', year]].sort_values(by=year, ascending=False).head(5)
        plt.figure(figsize=(10, 6))
        plt.barh(top_10['Country Name'], top_10[year], color='green')
        plt.xlabel(f'Population in {year}')
        plt.title(f'Top 10 Countries by Population in {year}')
        plt.gca().invert_yaxis()  
        plt.show()
```

```python
top_countries([1960 , 1980 , 2000 , 2023])
```

### Lowest 5 Countries by 1960 1980 2000 2020 's population

```python
def lowest_countries(list_years) :
    for year in list_years : 
        top_10 = dataset[['Country Name', year]].sort_values(by=year, ascending=True).head(5)
        plt.figure(figsize=(10, 6))
        plt.barh(top_10['Country Name'], top_10[year], color='red')
        plt.xlabel(f'Population in {year}')
        plt.title(f'Lowest 10 Countries by Population in {year}')
        plt.gca().invert_yaxis()  
        plt.show()
```

```python
lowest_countries([1960 , 1980 , 2000 , 2023])
```

# Algeria Population

```python
population_algeria = dataset[dataset['Country Name'] == 'Algeria']
```

```python
population_algeria
```

```python

years = population_algeria.columns[1:].tolist()
population_algeria_values = population_algeria.iloc[0, 1:].values
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(years, population_algeria_values, width=0.5, color='red')
ax.set_xticks(np.arange(1960, 2024, 10))
ax.set_ylabel('population')
ax.set_title('Algeria population')
plt.show()
```
